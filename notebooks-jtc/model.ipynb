```python
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
master_df = pd.read_csv('master_df.csv')
```

```python
master_df.columns
```

```python
hd_df = master_df[[
                    'N_day', 'N_day_plus','AcresOnLand','AcresOnLandPlus',
                   'HD_Store_Count',
                   'HD_Store_CountPlus'
                   ]]
                   #'hd_5days_avg','hd_10day_avg']]
hd_y = master_df['hd_rel_close']

X2 = sm.add_constant(hd_df)
est = sm.OLS(hd_y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
from sklearn.ensemble import RandomForestRegressor
regre = RandomForestRegressor(max_depth=2, random_state=0)
regre.fit(hd_df, hd_y)

```

```python
(pd.DataFrame({'features': list(hd_df.columns),
              'importance': regre.feature_importances_}).sort_values(by=['importance'],ascending=False)
                                                        .reset_index(drop=True)).to_clipboard()
```

```python
lowes_df = master_df[['N_day', 'N_day_plus',
                   'AcresOnLand','AcresOnLandPlus',
                   'LOW_Store_Count','LOW_Store_CountPlus']]
lowes_y = master_df['lowes_rel_close']

X2 = sm.add_constant(lowes_df)
est = sm.OLS(lowes_y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
regre = RandomForestRegressor(max_depth=2, random_state=0)
regre.fit(lowes_df, lowes_y)
```

```python
(pd.DataFrame({'features': list(lowes_df.columns),
              'importance': regre.feature_importances_}).sort_values(by=['importance'],ascending=False)
                                                        .reset_index(drop=True)).to_clipboard()
```
